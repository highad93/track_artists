```python
from time import sleep

chunk = 50
calls = 0


artist_id = []
for start in range(0, len(df), chunk):
  end = start + chunk
  ids = [str(x) for x in df.song_id[start:end].values]
  ids = ",".join(ids)
  response = requests.get(f"https://api.spotify.com/v1/tracks?ids={ids}", headers = headers).json()

  calls += 1
  if calls == 150:
    sleep(30)
    calls = 0

  for track in response["tracks"]:
    try:
      artist_id.append(track["artists"][0]["id"])
    except TypeError: #TypeError None type not subscriptable
      continue


artist_id = set(artist_id)
```
